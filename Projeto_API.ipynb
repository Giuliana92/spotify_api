
# Buscando o Artista a ser analisado

```python
imagine_dragons = 'spotify:artist:53XhwfbYqKCa1cC15pYq2q'
```

```python
spotify = spotipy.Spotify(client_credentials_manager=scc())
```

```python
results = spotify.artist_top_tracks(imagine_dragons)
```

## Buscando as 10 músicas mais tocadas do artista

```python
# Buscando as 10 músicas mais tocadas

for track in results['tracks'][:10]:
    print('track    : ' + track['name'])
    print (track ['preview_url'])
    print('cover art: ' + track['album']['images'][0]['url'])
    print()
```

# Buscando dados dos Álbuns

```python
results = spotify.artist_albums(imagine_dragons, album_type='album')
albums = results['items']
while results['next']:
    results = spotify.next(results)
    albums.extend(results['items'])

for album in albums:
    print(album['name'])
```

```python
Night_Visions = 'spotify:album:6nxDQi0FeEwccEPJeNySoS'
Smoke_Mirrors = 'spotify:album:0Ikw6ho559687KCPbSjr0K'
Envolve = 'spotify:album:33pt9HBdGlAbRGBHQgsZsU'
Origins = 'spotify:album:3JfSxDfmwS5OeHPwLSkrfr'
```

## Bucando dados do Álbum: Night Visions

```python
# Aplicando 'for' para percorrer o álbum e obter as tracks

Night_Visions_Tracks=[]
for i in range (len(spotify.album_tracks( Night_Visions )['items'])):
    Night_Visions_Tracks.append(spotify.album_tracks( Night_Visions )['items'][i]['uri'])
```

```python
Night_Visions_Tracks
```

```python
# Obter recursos de áudio para as faixas com base em seus parâmetros de IDs do Spotify

spotify.audio_features( Night_Visions_Tracks )
```

```python
Musicas_Night_Visions = spotify.audio_features( Night_Visions_Tracks )
```

```python
df_musica_night_visions = pd.json_normalize(Musicas_Night_Visions)
```

```python
# Criando nova coluna com o nome do Álbum

df_musica_night_visions ['Name_Album']='NightVisions'
```

```python
df_musica_night_visions
```

```python
# Buscando os dados do Álbum

Tracks_Night_Visions = spotify.album_tracks (Night_Visions)
```

```python
df_night_visions = pd.json_normalize(Tracks_Night_Visions ['items'])
```

```python
df_night_visions = df_night_visions.drop ('artists', axis=1)
```

```python
df_night_visions = df_night_visions.drop ('preview_url', axis=1)
```

```python
# Criando nova coluna com o nome do Álbum

df_night_visions ['Name_Album']='NightVisions'
```

```python
df_night_visions
```
